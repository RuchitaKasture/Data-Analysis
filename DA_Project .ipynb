```python
plt.figure(figsize=(12,5))
plt.scatter(placment_cell["hsc_p"],placment_cell["degree_t"],marker="^", cmap="plasma")
plt.xlabel("hsc_p")
plt.ylabel("degree_t")
plt.title("Marks")
plt.colorbar(label="hsc_p")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 4))
plt.hist(placment_cell["degree_t"], bins=20, color="red")
plt.xlabel("status")
plt.xticks(range(10, 38, 80))
plt.show()
```

```python

plt.pie(placment_cell["hsc_p"], labels=placment_cell["status"],
        colors=["#cc0000","#ff0000","#ff3333","#ff6666","#ff9999","#ffcccc"])
plt.title("specialisation")
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.bar(placment_cell["gender"],placment_cell["salary"] , width=1, edgecolor="black",
        color=["#cc0000","#ff0000","#ff3333","#ff6666","#ff9999","#ffcccc"])
plt.xlabel("gender")
plt.ylabel("salary")
plt.xticks(range(2,18,9))
plt.title("Salary of Male and Female ")
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.boxplot(placment_cell["degree_p"], vert=False, meanline=True, notch=True, showmeans=True)
plt.xlabel("Degree")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.lineplot(placment_cell["degree_t"])
plt.title("Total Degree Type")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.lineplot(placment_cell, x="sl_no", y="degree_p", hue="status")
plt.title("Total Degree Type")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(placment_cell, x="degree_p", hue="gender")
plt.title("Count of Student Pass In degree")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.scatterplot(placment_cell, x="salary", y="specialisation", hue="gender")
plt.title("working member with spcialisation and the salry")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.kdeplot(placment_cell, x="etest_p",fill= True, hue="status", palette="rocket")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(placment_cell, x="ssc_p", bins=30, kde=True, hue="ssc_b")
plt.xticks(range(40, 90))
plt.show()
```

```python
sns.jointplot(placment_cell, x="ssc_p", y="degree_p", height=4, kind="hex")
plt.show()
```

```python
plt.figure(figsize=(10,4),facecolor="grey")
sns.violinplot(placment_cell, x="degree_p", y="degree_t", palette="mako", fill=True, )
plt.xticks(range(40,90))
plt.title("Gender By degree Analysis")
plt.show()
```

```python
sns.pairplot(placment_cell.drop("sl_no",axis=1),kind="kde", hue="status", palette="rocket")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.catplot(placment_cell, x="status", y="gender",height=4)
plt.show()
```
